```python
class Application:
    def __init__(self, name, position, status):
        self.name = name
        self.position = position
        self.status = status

class ApplicationTracker:
    def __init__(self):
        self.applications = []

    def add_application(self, name, position, status):
        application = Application(name, position, status)
        self.applications.append(application)

    def update_status(self, name, new_status):
        for application in self.applications:
            if application.name == name:
                application.status = new_status
                break

    def list_applications(self):
        for application in self.applications:
            print(f"Name: {application.name}, Position: {application.position}, Status: {application.status}")

    def save_to_file(self, filename):
        with open(filename, "w") as file:
            for application in self.applications:
                file.write(f"{application.name},{application.position},{application.status}\n")

    def load_from_file(self, filename):
        self.applications = []
        try:
            with open(filename, "r") as file:
                lines = file.readlines()
                for line in lines:
                    data = line.strip().split(',')
                    if len(data) == 3:
                        name, position, status = data
                        self.add_application(name, position, status)
        except FileNotFoundError:
            print("File not found.")

if __name__ == "__main__":
    tracker = ApplicationTracker()

    while True:
        print("\nApplication Tracking System")
        print("1. Add Application")
        print("2. Update Application Status")
        print("3. List Applications")
        print("4. Save to File")
        print("5. Load from File")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter applicant's name: ")
            position = input("Enter position applied for: ")
            status = input("Enter application status: ")
            tracker.add_application(name, position, status)
            print("Application added successfully!")

        elif choice == "2":
            name = input("Enter applicant's name: ")
            new_status = input("Enter new application status: ")
            tracker.update_status(name, new_status)
            print("Application status updated!")

        elif choice == "3":
            tracker.list_applications()

        elif choice == "4":
            filename = input("Enter filename to save data: ")
            tracker.save_to_file(filename)
            print("Data saved to file!")

        elif choice == "5":
            filename = input("Enter filename to load data from: ")
            tracker.load_from_file(filename)
            print("Data loaded from file!")

        elif choice == "6":
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please try again.")
```
