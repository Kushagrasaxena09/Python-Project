```python
import math
import time

class Rocket:
    def __init__(self, name, thrust, mass):
        self.name = name
        self.thrust = thrust  # Thrust force in Newtons
        self.mass = mass      # Mass in kilograms
        self.velocity = 0.0  # Initial velocity
        self.altitude = 0.0  # Initial altitude

    def launch(self):
        print(f"Launching {self.name}!")
        time.sleep(1)

        time_step = 1  # Time step for simulation in seconds
        for t in range(0, 100, time_step):
            # Calculate acceleration due to thrust and gravity
            acceleration = (self.thrust / self.mass) - 9.8  # 9.8 m/s^2 is Earth's gravity

            # Update velocity and altitude using kinematic equations
            self.velocity += acceleration * time_step
            self.altitude += self.velocity * time_step

            # Print status
            print(f"Time: {t} seconds - Altitude: {self.altitude:.2f} meters - Velocity: {self.velocity:.2f} m/s")

        print(f"{self.name} has reached an altitude of {self.altitude:.2f} meters.")

def main():
    print("Rocket Launch Simulation")

    # Create a rocket
    rocket_name = "Suryayan 2.0"
    thrust_force = 8.2e6  # Approximate thrust force of Falcon 9 in Newtons
    rocket_mass = 549054  # Approximate mass of Falcon 9 in kilograms
    falcon9 = Rocket(rocket_name, thrust_force, rocket_mass)

    # Launch the rocket
    falcon9.launch()

if __name__ == "__main__":
    main()
```
