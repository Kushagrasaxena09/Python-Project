```python
# Import necessary libraries
import tkinter as tk
from tkinter import messagebox

# Define a dictionary of symptoms and corresponding medicines
symptoms_to_medicines = {
    'headache': 'Aspirin',
    'fever': 'Paracetamol',
    'cough': 'Cough syrup',
    'sore throat': 'Lozenges',
    'blood pressure':'Telmisartin'
    
    # Add more symptoms and medicines as needed
}

# Create a function to suggest medicines based on symptoms
def suggest_medicine():
    # Get the user's input from the text entry field
    user_symptoms = entry.get().lower()

    # Split user input into individual symptoms
    user_symptoms_list = user_symptoms.split(',')

    # Initialize a list to store suggested medicines
    suggested_medicines = []

    # Iterate through user symptoms and suggest medicines
    for symptom in user_symptoms_list:
        symptom = symptom.strip()
        if symptom in symptoms_to_medicines:
            suggested_medicines.append(symptoms_to_medicines[symptom])
        else:
            suggested_medicines.append(f"No medicine suggestion for '{symptom}'")

    # Display the suggested medicines in a message box
    suggested_medicines_str = '\n'.join(suggested_medicines)
    messagebox.showinfo("Medicine Suggestion", suggested_medicines_str)

# Create a GUI window
app = tk.Tk()
app.title("Medicine Suggestion App")

# Create a label
label = tk.Label(app, text="Enter your symptoms (comma-separated):")
label.pack()

# Create a text entry field
entry = tk.Entry(app)
entry.pack()

# Create a button to trigger medicine suggestion
button = tk.Button(app, text="Get Medicine", command=suggest_medicine)
button.pack()

# Start the GUI application
app.mainloop()
```
