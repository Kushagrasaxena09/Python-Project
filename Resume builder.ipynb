```python
class ResumeBuilder:
    def __init__(self):
        self.name = ""
        self.email = ""
        self.phone = ""
        self.education = []
        self.experience1 = []
        self.experience2 = []
        self.internship=[]
        self.skills = []
    
    def get_personal_info(self):
        print("Enter your personal information:")
        self.name = input("Name: ")
        self.email = input("Email: ")
        self.phone = input("Phone: ")

    def add_education(self):
        print("Enter your education information:")
        degree = input("Degree: ")
        school = input("School/University: ")
        year = input("Year of Graduation: ")
        self.education.append((degree, school, year))

    def add_experience1(self):
        print("Enter your work experience 1:")
        title = input("Job Title: ")
        company = input("Company: ")
        duration = input("Duration (e.g., 2 years): ")
        self.experience1.append((title, company, duration))
    def add_experience2(self):
        print("Enter your work experience 2:")
        title = input("Job Title: ")
        company = input("Company: ")
        duration = input("Duration (e.g., 2 years): ")
        self.experience2.append((title, company, duration))    

    def add_skills(self):
        print("Enter your skills (comma-separated):")
        skills_input = input()
        self.skills = skills_input.split(",")
    def add_internship(self):
        print("Enter your Internship:")
        title = input("Job Title: ")
        company = input("Company: ")
        duration = input("Duration (e.g., 2 years): ")
        self.internship.append((title, company, duration))   

    def generate_resume(self):
        resume = f"Name: {self.name}\n"
        resume += f"Email: {self.email}\n"
        resume += f"Phone: {self.phone}\n\n"

        resume += "Education:\n"
        for degree, school, year in self.education:
            resume += f"- {degree} in {year} from {school}\n"
        
        resume += "\nWork Experience1:\n"
        for title, company, duration in self.experience1:
            resume += f"- {title} at {company} ({duration})\n"
        resume += "\nWork Experience2:\n"
        for title, company, duration in self.experience2:
            resume += f"- {title} at {company} ({duration})\n"    
        resume += "\nWork Internship:\n"
        for title, company, duration in self.internship:
            resume += f"- {title} at {company} ({duration})\n"
        resume += "\nSkills:\n"
        resume += ", ".join(self.skills)

        return resume

def main():
    print("Welcome to the Resume Builder App!")
    builder = ResumeBuilder()

    builder.get_personal_info()
    builder.add_education()
    builder.add_experience1()
    builder.add_experience2()
    builder.add_skills()
    builder.add_internship()

    resume_text = builder.generate_resume()

    print("\nGenerated Resume:\n")
    print(resume_text)

if __name__ == "__main__":
    main()
```
