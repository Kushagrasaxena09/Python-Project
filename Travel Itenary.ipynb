```python
class TravelItinerary:
    def __init__(self):
        self.itinerary = []

    def add_destination(self, destination, date, activities):
        self.itinerary.append({"destination": destination, "date": date, "activities": activities})

    def view_itinerary(self):
        if not self.itinerary:
            print("Your travel itinerary is empty.")
        else:
            print("\nTravel Itinerary:")
            for idx, trip in enumerate(self.itinerary, start=1):
                print(f"Trip {idx}:")
                print(f"Destination: {trip['destination']}")
                print(f"Date: {trip['date']}")
                print("Activities:")
                for activity in trip['activities']:
                    print(f"  - {activity}")
                print()

def main():
    travel_planner = TravelItinerary()

    while True:
        print("\nTravel Itinerary App")
        print("1. Add Trip")
        print("2. View Itinerary")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            destination = input("Enter destination: ")
            date = input("Enter date: ")
            activities = []

            while True:
                activity = input("Enter activity (or leave blank to finish): ")
                if not activity:
                    break
                activities.append(activity)

            travel_planner.add_destination(destination, date, activities)
            print("Trip added to itinerary.")

        elif choice == "2":
            travel_planner.view_itinerary()

        elif choice == "3":
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
